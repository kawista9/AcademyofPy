```python
#Dpendencies and Setup

import pandas as pd
import numpy as np
```

```python
#File to Load
school_data_to_load ="schools_complete.csv"
students_data_to_load="students_complete.csv"

school_data=pd.read_csv(school_data_to_load)
student_data = pd.read_csv(students_data_to_load)

school_data_complete=pd.merge(student_data, school_data, how="left", on=["school_name","school_name"])
```

```python

#converting dictioning into a data frame; over of the data
df_schoolData=pd.DataFrame(school_data_complete)
df_schoolData.describe()
```

```python
organized_schoolData_df=df_schoolData[["school_name", "School ID","type","budget", "size",
                                      "grade", "gender", "student_name", "Student ID", "math_score","reading_score"]]

renamed_schoolData_df=organized_schoolData_df.rename(columns={"school_name":"School Name", "type":"Type","budget":"Budget",
                                                  "size":"Size","grade":"Grade Level", "gender":"Gender",
                                                  "student_name":"Student Name","math_score":"Math Score", "reading_score":"Reading Score"})
reducedschooldata_df=renamed_schoolData_df[["School Name","School ID", "Budget", "Size","Grade Level", "Gender", "Student Name","Student ID", "Math Score", "Reading Score"]]
reducedschooldata_df.head()


```

```python
#school summary

#find the unique budget values and sum those values
totalbudget=sum(reducedschooldata_df["Budget"].unique())

#find how many schools there are in the data  set
schoolcount=len(reducedschooldata_df["School Name"].unique())

#find the total number of students across the district in this data set
totalstudents=len(reducedschooldata_df["Student ID"].unique())

#average math and reading scores
avgmath=((sum(reducedschooldata_df["Math Score"]))/(len(reducedschooldata_df["Math Score"])))


avgreading=((sum(reducedschooldata_df["Reading Score"]))/(len(reducedschooldata_df["Reading Score"])))


#passing rate for math
mathscoresfiltered = reducedschooldata_df["Math Score"]>=70
passingmathscores=mathscoresfiltered.loc[mathscoresfiltered]
percent_passingmathscores=(len(passingmathscores)/totalstudents)*100

#passing rate for reading
readingscoresfiltered=reducedschooldata_df["Reading Score"]>=70
passingreadingscores=readingscoresfiltered.loc[readingscoresfiltered]
percent_passingreadingscores=(len(passingreadingscores)/totalstudents)*100

#passing rate for both
percent_passingboth=(percent_passingmathscores + percent_passingreadingscores)/2
percent_passingboth

summarytable = pd.DataFrame({"Total Schools" : schoolcount,
                                    "Total Students":[totalstudents],
                                    "Total Budget":[totalbudget],
                                     "Average Math Score":[avgmath],
                                     "Average Reading Score":[avgreading],
                                    "% Passing Math":[percent_passingmathscores],
                                    "% Passing Reading":[percent_passingreadingscores],
                                    "Overall Passing Rate":percent_passingboth})

summarytable




```
